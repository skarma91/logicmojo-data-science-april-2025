#### Median

```python
def median(X):
    values = sorted(X)
    n = len(values)
    if n % 2 != 0:
        return values[(n-1)//2]
    else:
        return 0.5*(values[(n-1)//2] + values[(n+1)//2])
```

#### Covariance

```python
def covariance(X, Y):
    x_bar = X.mean()
    y_bar = Y.mean()
    n = X.shape[0]
    cov = ((X - x_bar) @ (Y - y_bar)) / n
    return cov
```

```python
covariance(sample_data['Height'].values, sample_data['Weight'].values)
```

```python
covariance(data['Height'].values, data['Weight'].values)
```

```python
np.cov(data['Height'], data['Weight'])
```

#### Correlation

```python
def correlation(X, Y):
    cov = covariance(X, Y)
    std_x = X.std()
    std_y = Y.std()
    corr = cov / (std_x * std_y)
    return corr
```

```python
correlation(sample_data['Height'], sample_data['Weight'])
```

```python
correlation(data['Height'], data['Weight'])
```
