### Helper function to plot the histogram

```python
import matplotlib.pyplot as plt

def plot_hist(x, edgecolor='k', nbins=20):
    plt.figure(figsize=(10,8))
    plt.hist(x, edgecolor=edgecolor, bins=nbins)
    plt.xlabel('Values')
    plt.ylabel('Frequency')
    plt.show()
```

```python
### uniformly distributed random values

values = np.random.rand(100000)

plot_hist(values, 'k', 20)
```

```python
values = np.random.uniform(2, 6, 100000)

plot_hist(values)
```

```python
### Normally distributed random variable

values = np.random.randn(10000)

plot_hist(values)
```

```python
### mean = 5, std dev = 2

values = np.random.normal(loc=5, scale=2, size=10000)

plot_hist(values)
```

```python
def plot_bar(x):

    from collections import Counter

    count_dict = Counter(x)

    sorted_counts = sorted(count_dict.items())

    values = [v[0] for v in sorted_counts]
    counts = [v[1] for v in sorted_counts]

    plt.figure(figsize=(10,8))
    plt.bar(x=values, height=counts)
    plt.show()
```

```python
values = np.random.randint(1, 50, 10000)

plot_bar(values)
```

### Simulating random experiments

#### Simulate coin toss

```python
### Unbiased coin (equal probabilities of getting head and tail)

### Biased coin (head with probability p and tail with probaility (1-p))

def simulate_cointoss(p = 0.5):

    if p > 1.0 or p < 0:
        raise ValueError("The probability p should be between 0 to 1")
    
    v = np.random.rand(1)[0]

    if v <= p:
        return 'H'
    else:
        return 'T'    
```

```python
def multiple_cointoss(n, p=0.5):

    results = [simulate_cointoss(p) for i in range(n)]

    return results
```

```python
results = multiple_cointoss(1000, p=0.5)
```

```python
plot_bar(results)
```

```python
results = multiple_cointoss(1000, p=0.2)

plot_bar(results)
```

```python
results = multiple_cointoss(1000, p=0.7)

plot_bar(results)
```

### Cumulative frequency and Cumulative distribution

```python
values = np.random.normal(4, 2, 10000)
```

```python
plot_hist(values)
```

```python
n,bins,patches = plt.hist(values, bins=20)
```

```python
np.cumsum(n)
```

```python
plt.bar(x=np.arange(n.size), height=np.cumsum(n))
```

```python
plt.bar(x=np.arange(n.size), height=np.cumsum(n)/10000)
```

```python
plt.figure(figsize=(10,8))

plt.hist(values, bins=20, cumulative=True, edgecolor='k', color='c')
plt.hist(values, bins=20, cumulative=False, edgecolor='k', color='m')

plt.show()
```
