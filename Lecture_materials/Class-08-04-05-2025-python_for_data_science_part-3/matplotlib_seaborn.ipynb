## Matplotlib

- Matplotlib is one of the most popular Python packages used for data visualization.
- It is a cross-platform library for making 2D plots from data in arrays.
- It provides an object-oriented API that helps in embedding plots in applications.

install matplotlib in your virtual environment using 

```
pip install matplotlib
```

### Importing the necessary libraries

````python
import numpy as np
import matplotlib.pyplot as plt   # from matplotlib import pyplot as plt
````

### Creating data for plot

````python
x = np.linspace(0,4*np.pi,1000)
x.shape
````

````python
y = np.sin(x)
y.shape
````

### Line Plot

````python
plt.figure(figsize = (10,7))
plt.plot(x,y,'r')
plt.grid()
plt.xlabel('X')
plt.ylabel('sin(X)')
plt.title('sin(X) vs X')
plt.show()
````

### Simultaneously plotting multiple lineplots in the same figure

````python
y1 = np.sin(x)
y2 = np.cos(x)
````

````python
plt.figure(figsize = (10,6))
plt.plot(x,y1,c = 'b',label = 'sin')
plt.plot(x,y2,c = 'r',label = 'cos')
plt.grid()
plt.legend()
plt.xlabel('X')
plt.ylabel('sin or cos x')
plt.title('My_plots')
plt.show()
````

### Few more examples with line plot

````python
y1 = 2*y1
y2 = 3*y2
plt.figure(figsize = (8,5))
plt.plot(y1,y2)
plt.show()
````

### Subplots

![](https://scipy-lectures.org/_images/sphx_glr_plot_subplot-grid_thumb.png)

````python
x = np.linspace(1,10,1000)

y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.log(x)
y4 = np.exp(x)

plt.figure(figsize = (8,8))
plt.suptitle('My plots')

plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title("Sin(X)")

plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title('Cos(x)')

plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title('log(x)');

plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title('exp(x)');
````

````python
# Plot 6 subplots
# 1. sin(x)
# 2. cos(x)
# 3,4. log(x)
# 5,6. exp(x)

x = np.linspace(0.001,10,1000)
````

````python
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.log(x)
y4 = np.exp(x)
````

````python
plt.figure(figsize = (8,8))
plt.suptitle('My plots')

plt.subplot(3,2,1)
plt.plot(x,y1)
plt.title("Sin(X)")
plt.grid()

plt.subplot(3,2,2)
plt.plot(x,y2)
plt.title('Cos(x)')
plt.grid()

plt.subplot(3,2,(3,5))
plt.plot(x,y3)
plt.title('log(x)')
plt.grid()

plt.subplot(3,2,(4,6))
plt.plot(x,y4)
plt.title('exp(x)');
plt.grid()

plt.show()
````

### Bar Plot

````python
plt.text(x, y, 'text')
````

````python
def add_labels(x, y):
    for i in range(len(x)):
        plt.text(i, y[i]+0.5, y[i])
````

````python
values = np.array([10,15,5,19,23,17])

x = list('abcdef')

plt.bar(x, height = values, width=0.8)

add_labels(x, values)
````

````python
labels = ['Jan', 'Feb', 'Mar']

values = [100, 150, 90]

 

# Create bar chart

fig, ax = plt.subplots()

bars = ax.bar(labels, values)

 

# Add value labels on top of each bar

ax.bar_label(bars, label_type='center')
````

### 3D Scatter Plot

````python
from mpl_toolkits.mplot3d import Axes3D
````

````python
x = np.random.rand(30)
y = np.random.rand(30)
z = np.random.rand(30)
````

````python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(111, projection = '3d')
ax.scatter3D(x,y,z)
plt.show()
````

### 3D surface plot

````python
(x1,y1) = np.meshgrid(x,y)
````

````python
z1 = x1**2 + y1**2 - 45
````

````python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(111, projection = '3d')
ax.plot_surface(x1,y1,z1)
plt.show()
````

## Seaborn

- Seaborn is an amazing data visualization library for statistical graphics plotting in Python.
- It provides beautiful default styles and colour palettes to make statistical plots more attractive.
- It is built on the top of the matplotlib library and also closely integrated to the data structures from pandas.


Install seaborn using 
```
pip install seaborn
```

### Load Data To Construct Seaborn Plots

There are some built-in datasets in seaborn, which comes very handy while practicing seaborn.

````python
import seaborn as sns
print(sns.get_dataset_names())
print(len(sns.get_dataset_names()))
````

````python
df = sns.load_dataset('car_crashes')
print(df.head())
````

### Styling and Themes in Seaborn

````python
plt.scatter(df.speeding,df.alcohol)
plt.show()
````

````python
plt.scatter(df.speeding,df.alcohol)
sns.set()
plt.show()
````

````python
plt.scatter(df.speeding,df.alcohol)
sns.set_style("whitegrid")
plt.show()
````

````python
plt.scatter(df.speeding,df.alcohol)
sns.set_style("dark")
plt.show()
````

````python
plt.scatter('speeding','alcohol',data=df)
sns.set_style("darkgrid", {'grid.color': '.2'})
sns.despine()
plt.show()
````

### Joint Plot

Joint Plot draws a plot of two variables with bivariate and univariate graphs. It uses the Scatter Plot and Histogram. Joint Plot can also display data using Kernel Density Estimate (KDE) and Hexagons.

````python
sns.set_style("dark")
tips=sns.load_dataset('tips')
tips.head()
````

````python
sns.jointplot(x='total_bill', y='tip',data=tips)
````

### Regplot

Regplot is one of the functions in Seaborn that are used to visualize the linear relationship as determined through regression. Also, youll see a slightly shaded portion around the regression line which indicates how much the pints are scattered around a certain area. Here are few of the examples

````python
sns.regplot(x="total_bill", y="tip", data=tips)
````

### Boxplot

````python
sns.boxplot(x="day", y="total_bill", data=tips)
````

### Pairplot

````python
df = sns.load_dataset('iris')
df.head()
````

````python
sns.set_style("ticks")
sns.pairplot(df,hue = 'species',diag_kind = "kde",kind = "scatter",palette = "husl")
plt.show()
````

### Countplot

````python
# load dataset
titanic = sns.load_dataset('titanic')
titanic.head()
````

````python
sns.set_context('paper')
sns.countplot(x = 'class', hue = 'who', data = titanic)
plt.title('Survivors')
plt.show()

# stacked bar graph
````
